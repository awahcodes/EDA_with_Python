```python
#import data with read and print

import pandas as pd
df = pd.read_csv('worldpopulation.csv')
print(df)
```

```python
#Viewing info on the dataset to see column, non-null and datatype overview

df.info()
```

```python
#describe() gives some stats overview of the data

df.describe()
```

```python
#isnull() shows the null and null value status as a bool

df.isnull()
```

```python
#to show to sum of the null 

df.isnull().sum()
```

```python
#to see the number of unique values in the data against the column
df.nunique()
```

```python
#sorting by a particular column. By default, we will have  the sort by ascendding order for column "2022 population"

df.sort_values('2022 Population')
```

```python
#To sort by descending order, we use ascending= "False"
df.sort_values('2022 Population', ascending = False)
```

```python
#To check the relationship btw my data, i use correlation "corr()"

df.corr()
```

```python
#To group by a parameter and show the mean values

df2 = df.groupby("Continent").mean()
print(df2)
```

```python
#Having the data on a chart
df2.plot()
```

```python
#Te above doesnt plot niely so i can transpose the axis
df3 = df2.transpose()
print(df3)
```

```python
#I will plot the new transpose axix
df3.plot()
```
